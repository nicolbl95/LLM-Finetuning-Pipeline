```python
from datasets import load_dataset
import pandas as pd

# Charger le dataset depuis Hugging Face
dataset = load_dataset("gbharti/finance-alpaca")

# Afficher la structure générale
print("Structure du dataset :")
print(dataset)

# Afficher les splits disponibles
print("\nSplits disponibles :")
print(dataset.keys())

# Compter les exemples dans chaque split
print("\nNombre d'exemples :")
for split_name in dataset.keys():
    print(f"{split_name}: {len(dataset[split_name])} exemples")

# Afficher le premier exemple complet
print("\nPremier exemple complet :")
print(dataset["train"][0])

# Afficher quelques exemples lisibles
for i in range(3):
    ex = dataset["train"][i]
    
    print(f"\n--- Exemple {i+1} ---")
    print("Instruction :")
    print(ex.get("instruction", "")[:500])
    
    print("\nInput :")
    print(ex.get("input", "")[:500])
    
    print("\nRéponse :")
    print(ex.get("output", "")[:500])

# Convertir en DataFrame pour explorer
df = pd.DataFrame(dataset["train"])

print("\nColonnes du dataset :")
print(df.columns)

print("\nValeurs manquantes :")
print(df.isnull().sum())

# Longueur moyenne des réponses
df["output_length"] = df["output"].astype(str).apply(len)

print("\nStatistiques sur la longueur des réponses :")
print(df["output_length"].describe())

df.head()
```
